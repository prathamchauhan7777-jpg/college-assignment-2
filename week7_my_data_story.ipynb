```python
# Step 1: Specify the file to be analyzed
data_file = "my_data.txt"

# Step 2: Load the file content into a list
with open(data_file, "r") as f:
    content_lines = f.readlines()

# Step 3: Display the contents of the file
print("=== CONTENT OF MY DATA FILE ===")
for entry in content_lines:
    print(entry.strip())

# Step 4: Analyze the file for basic statistics
line_count = len(content_lines)
word_count = sum(len(line.split()) for line in content_lines)
max_length_line = max(content_lines, key=len).strip()
first_entry = content_lines[0].strip()
last_entry = content_lines[-1].strip()
char_count = sum(len(line) for line in content_lines)

# Step 5: Present the results
print("\n--- FILE ANALYSIS SUMMARY ---")
print(f"Total lines: {line_count}")
print(f"Total words: {word_count}")
print(f"Longest line: {max_length_line}")
print(f"First line: {first_entry}")
print(f"Last line: {last_entry}")
print(f"Total characters: {char_count}")
```
