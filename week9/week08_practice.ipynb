## Practice Exercise 2.1: Spending Trend Analysis

::: {.callout-tip}
## Business Context

You're helping a friend understand their spending habits over time.

**Challenge**: Ask AI to help you:
1. Convert dates to datetime format
2. Analyze spending trends by day of week
3. Identify if spending has increased over time
4. Spot any unusual spending patterns

**Focus**: Practice asking AI to explain any complex date handling code it provides.
:::

**Your Analysis:**

````python
You are my data assistant helping me analyze personal spending habits over time.
The dataset includes transaction dates and amounts.

Please:
1. Convert the Date column to datetime format and extract useful features like day of week and month.
2. Calculate total spending by day of week to find when I spend the most.
3. Analyze whether spending is increasing over time â€” perhaps by comparing monthly averages.
4. Detect any unusual spending patterns (e.g., outliers or sudden spikes).
5. Provide a simple visualization and explain any complex pandas or datetime operations in plain language.
6. Summarize findings clearly for a non-technical audience.
import pandas as pd
import matplotlib.pyplot as plt

# Load and prepare data
df = pd.read_csv("transactions.csv")
df['Date'] = pd.to_datetime(df['Date'])

# Extract useful time features
df['Day_of_Week'] = df['Date'].dt.day_name()
df['Month'] = df['Date'].dt.to_period('M')

# Summarize spending patterns
spend_by_day = df.groupby('Day_of_Week')['Amount'].sum().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Spending trend over time
monthly_trend = df.groupby('Month')['Amount'].sum()

# Detect outliers (very high spend days)
daily_totals = df.groupby('Date')['Amount'].sum()
threshold = daily_totals.mean() + 2 * daily_totals.std()
unusual_days = daily_totals[daily_totals > threshold]

# Plot
monthly_trend.plot(kind='line', title='Monthly Spending Trend', ylabel='Total Amount')
plt.show()

Step 2: Insights

You spend most often on weekends, especially Saturday.

Spending has gradually increased over the past few months, possibly reflecting higher leisure or discretionary purchases.

Unusual spikes on a few dates suggest larger one-time expenses â€” these may be big-ticket items or special events.

Step 3: Explanation

pd.to_datetime() converts text-based dates into pandasâ€™ datetime format, enabling time-based grouping and trend analysis.

df['Date'].dt.day_name() extracts the weekday names directly from each date.

The use of groupby() lets us easily summarize data by any time dimension (day, month, etc.).

My reflections

What did AI do well?
It used clear, interpretable pandas code and provided both analysis and business-friendly explanations.

What could AI improve?
It could label charts more clearly and automatically highlight the unusual days in the plot.

Did it answer the business question?
Yes : it identified spending patterns over time, days with higher spending, and potential anomalies.
````

## Practice Exercise 2.2: Budget Category Analysis

::: {.callout-tip}
## Personal Finance Scenario

Someone wants to set a realistic monthly budget based on their spending history.

**Challenge**: Use AI to help:
1. Calculate average monthly spending by category
2. Identify categories with high variation (inconsistent spending)
3. Suggest realistic budget amounts for each category
4. Find categories where spending might be reduced

**Advanced**: Ask AI to explain the statistical concepts it uses.
:::

**Your Budget Analysis:**

````python
You are my data analysis assistant helping me understand spending habits for better budgeting.
I have a dataset of transactions with columns like: Date, Category, and Amount.

Please:
1. Convert the Date column to datetime format and extract month/year.
2. Calculate **average monthly spending by category**.
3. Identify **categories with high variation** (using standard deviation or coefficient of variation).
4. Suggest **realistic monthly budget targets** â€” slightly above the average for stable categories, and more conservative for volatile ones.
5. Highlight **categories where spending could be reduced**.
6. Explain any statistical terms or methods (like standard deviation or variance) in simple language.
7. Present results in a **clear summary table** and include **budget recommendations**.

AI response-

import pandas as pd

df = pd.read_csv("transactions.csv")
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M')

# Monthly spending by category
monthly_spend = df.groupby(['Month', 'Category'])['Amount'].sum().reset_index()

# Average and variation by category
budget_summary = monthly_spend.groupby('Category')['Amount'].agg(
    Average_Monthly_Spend='mean',
    Std_Deviation='std'
).reset_index()

# Add coefficient of variation (CV) to measure consistency
budget_summary['Coeff_of_Variation'] = (
    budget_summary['Std_Deviation'] / budget_summary['Average_Monthly_Spend']
).round(2)

# Suggest budget (slightly above average for stable categories)
budget_summary['Suggested_Budget'] = budget_summary.apply(
    lambda row: row['Average_Monthly_Spend'] * (1.1 if row['Coeff_of_Variation'] < 0.2 else 0.9),
    axis=1
).round(2)

budget_summary.sort_values(by='Average_Monthly_Spend', ascending=False)
Step 2: Results Example

Category	Avg Monthly Spend	Std Deviation	CV	Suggested Budget
Dining Out	$480	$95	0.20	$528
Groceries	$450	$60	0.13	$495
Entertainment	$220	$120	0.55	$198
Transportation	$160	$40	0.25	$144
Subscriptions	$95	$5	0.05	$105

Step 3: Insights & Recommendations

Dining Out and Groceries are the largest monthly expenses. These categories should get the highest budget priority.

Entertainment spending is highly variable (CV = 0.55), suggesting inconsistent habits â€” a reduction target of 10â€“15% could stabilize spending.

Subscriptions are predictable and low â€” safe to keep as-is.

Transportation shows moderate variation; budgeting slightly under the average encourages efficiency.

Step 4: Concept Explanation

Standard Deviation measures how spread out spending is from the average. A higher number means more unpredictable spending.

Coefficient of Variation (CV) shows variability relative to the mean â€” a CV above 0.3 often signals inconsistent spending habits.

Setting a budget slightly above or below the average helps create realistic goals that fit real spending behavior.
````
